## Task 1 Code: Output in combined_text.txt file

```python
csv_files = {
    'CSV1.csv': 'SHORT-TEXT',
    'CSV2.csv': 'TEXT',
    'CSV3.csv': 'TEXT',
    'CSV4.csv': 'TEXT'
}

output_file = 'combined_texts.txt'

with open(output_file, 'w') as outfile:
    outfile.write('')

for file, text_column in csv_files.items():
    df = pd.read_csv(file)

    text_column_data = df[text_column]

    with open(output_file, 'a') as outfile:
        for text in text_column_data:
            outfile.write(str(text) + '\n')

print(f"Text extraction completed. All text saved to {output_file}.")
```

```python
import spacy
import scispacy
from collections import Counter
import csv

# Load the SciSpaCy model
nlp = spacy.load("en_core_sci_sm")

# Read the contents of the file into the 'text' variable
with open('/content/combined_texts.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Function to clean and tokenize text
def tokenize_text(text):
    # Tokenize and clean the text
    doc = nlp(text)
    words = [token.text.lower() for token in doc if token.is_alpha]  # Only keep alphabetic tokens
    return words

# Initialize a Counter to aggregate word counts
total_word_counts = Counter()

# Process text in chunks
chunk_size = 1000000  # 1,000,000 characters per chunk
text_chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

for chunk in text_chunks:
    words = tokenize_text(chunk)
    total_word_counts.update(words)

# Get the top 30 most common words
top_30_words = total_word_counts.most_common(30)

# Write the top 30 words and their counts to a CSV file
with open('/content/top_30_words.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Word', 'Count'])  # Write the header
    writer.writerows(top_30_words)  # Write the top 30 words and their counts
```

```python
from transformers import AutoTokenizer
from collections import Counter
import csv

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Function to read, tokenize, and count unique tokens
def process_text(file_path, chunk_size=1000000):
    total_token_counts = Counter()

    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()

    # Split text into chunks
    text_chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

    for chunk in text_chunks:
        # Tokenize the text chunk
        tokens = tokenizer.tokenize(chunk)
        total_token_counts.update(tokens)

    # Get the top 30 most common tokens
    top_30_tokens = total_token_counts.most_common(30)

    return top_30_tokens

# Function to write top 30 tokens to a CSV file
def write_to_csv(top_tokens, output_path):
    with open(output_path, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Token', 'Count'])  # Write the header
        writer.writerows(top_tokens)  # Write the top tokens and their counts

# Paths to your input and output files
input_file_path = '/content/combined_texts.txt'
output_file_path = '/content/top_30_tokens_transformer.csv'

# Process the text and get the top 30 tokens
top_tokens = process_text(input_file_path)

# Write the results to a CSV file
write_to_csv(top_tokens, output_file_path)
```

```python
import spacy
import scispacy
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
from collections import Counter

# Load SciSpaCy models
nlp_sci = spacy.load("en_core_sci_sm")
ner_sci = spacy.load("en_ner_bc5cdr_md")

# Load BioBERT model using transformers
tokenizer_bert = AutoTokenizer.from_pretrained("dmis-lab/biobert-v1.1")
model_bert = AutoModelForTokenClassification.from_pretrained("dmis-lab/biobert-v1.1")
nlp_bert = pipeline("ner", model=model_bert, tokenizer=tokenizer_bert)

# Function to extract entities using SciSpaCy
def extract_entities_sci(text):
    doc = nlp_sci(text)
    diseases = [ent.text for ent in doc.ents if ent.label_ == "DISEASE"]
    drugs = [ent.text for ent in doc.ents if ent.label_ == "DRUG"]
    return diseases, drugs

# Function to extract entities using BioBERT
def extract_entities_bert(text):
    entities = nlp_bert(text)
    diseases = [ent['word'] for ent in entities if ent['entity'] == "B-DISEASE"]
    drugs = [ent['word'] for ent in entities if ent['entity'] == "B-DRUG"]
    return diseases, drugs

# Read the contents of the file into the 'text' variable
with open('/content/combined_texts.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Extract entities using both models
diseases_sci, drugs_sci = extract_entities_sci(text)
diseases_bert, drugs_bert = extract_entities_bert(text)

# Count entities
count_diseases_sci = Counter(diseases_sci)
count_drugs_sci = Counter(drugs_sci)
count_diseases_bert = Counter(diseases_bert)
count_drugs_bert = Counter(drugs_bert)

# Compare total counts
total_diseases_sci = len(diseases_sci)
total_drugs_sci = len(drugs_sci)
total_diseases_bert = len(diseases_bert)
total_drugs_bert = len(drugs_bert)

print("SciSpaCy Diseases Count:", total_diseases_sci)
print("BioBERT Diseases Count:", total_diseases_bert)
print("Difference in Diseases Count:", abs(total_diseases_sci - total_diseases_bert))

print("SciSpaCy Drugs Count:", total_drugs_sci)
print("BioBERT Drugs Count:", total_drugs_bert)
print("Difference in Drugs Count:", abs(total_drugs_sci - total_drugs_bert))

# Most common entities
common_diseases_sci = count_diseases_sci.most_common(10)
common_drugs_sci = count_drugs_sci.most_common(10)
common_diseases_bert = count_diseases_bert.most_common(10)
common_drugs_bert = count_drugs_bert.most_common(10)

print("Most Common Diseases (SciSpaCy):", common_diseases_sci)
print("Most Common Drugs (SciSpaCy):", common_drugs_sci)
print("Most Common Diseases (BioBERT):", common_diseases_bert)
print("Most Common Drugs (BioBERT):", common_drugs_bert)
```
